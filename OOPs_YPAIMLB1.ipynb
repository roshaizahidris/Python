```python
# OOPS
# OOP stands for Object-Oriented Programming.
# Object-oriented programming (OOP) is a programming paradigm based on the
# concept of "objects"

# Procedural programming is about writing procedures or methods that perform operations
# on the data, while object-oriented programming is about creating objects
# that contain both data and methods.

# Object-oriented programming has several advantages over procedural programming:

# a.) OOP is faster and easier to execute
# b.) OOP provides a clear structure for the programs
# c.) OOP helps to keep the Java/Python code DRY "Don't Repeat Yourself", and makes the code easier
#     to maintain, modify and debug
# d.) OOP makes it possible to create full reusable applications with less code and
#     shorter development time

# Class
# A class is a blueprint that defines the variables and the methods common to all objects.
# a class is a collection of methods and variables


# Object
# An object is called an instance of a class.


# Self Keyword
# The self parameter is a reference to the current instance of the class, and is used to
# access variables that belongs to the class.
# The first parameter of each method in a class must be the self ,
# which refers to the calling object. However, you can give any name to the first
# parameter, not necessarily self.

# __init__ in Python
#  Constructors are used to initializing the object’s state. The task of constructors is
# to initialize(assign values) to the data members of the class when
# an object of the class is created.
# The __init__() function is called automatically every time the class is being used to create
# a new object.
```

```python
# in order to access class variable or methods
# create the object of a class

class myclass:

  a = 10
  b = 20

  def func1():
    print('Hi I am a class function')

  def func2():
    print('Hi I am another class function')

# Creating An Object
# A class can have n number of objects
obj1 = myclass()
obj2 = myclass()

print(obj1.a)
print(obj2.a)
```

```python
# use self keyword
# first parameter of a class function must be SELF

# self represent the current calling instance (object)

class myclass:

  a = 10
  b = 20

  def func1(self):                           #  func1(obj1)      func1(obj2)
    print('Hi I am a class function')

  def func2(self):                               # func2(obj1)
    print('Hi I am another class function')


obj1 = myclass()
obj2 = myclass()

print(obj1.a)
print(obj2.a)

obj1.func1()
obj2.func1()

obj1.func2()
obj2.func2()
```

```python
# WAP to take 3 details from 3 users

# Taking details for user1
print("Enter details for User 1:")
user1_name = input("Name: ")
user1_email = input("Email: ")
user1_mobno = input("Mobile No: ")

# Taking details for user2
print("\nEnter details for User 2:")
user2_name = input("Name: ")
user2_email = input("Email: ")
user2_mobno = input("Mobile No: ")

# Taking details for user3
print("\nEnter details for User 3:")
user3_name = input("Name: ")
user3_email = input("Email: ")
user3_mobno = input("Mobile No: ")

# a) Print all information
print("\n--- All Users Information ---")
print("User1:", user1_name, user1_email, user1_mobno)
print("User2:", user2_name, user2_email, user2_mobno)
print("User3:", user3_name, user3_email, user3_mobno)

# b) Print information for 2nd user
print("\n--- Information for 2nd User ---")
print("Name:", user2_name)
print("Email:", user2_email)
print("Mobile No:", user2_mobno)

# c) Print mobile number of 3rd user
print("\n--- Mobile number of 3rd User ---")
print("Mobile No:", user3_mobno)
```

```python
# WAP to take 3 details (name, email, mobno) from 3 users using loops and lists

users = []  # list to store all users' details

# Taking details of 3 users
for i in range(3):
    print(f"\nEnter details for User {i+1}:")
    name = input("Name: ")
    email = input("Email: ")
    mobno = input("Mobile No: ")

    # store one user's info as a list inside the main list
    users.append([name, email, mobno])

# a) Print all information
print("\n--- All Users Information ---")
for i, user in enumerate(users, start=1):
    print(f"User{i}: Name = {user[0]}, Email = {user[1]}, Mobile No = {user[2]}")

# b) Print information for 2nd user
print("\n--- Information for 2nd User ---")
print(f"Name: {users[1][0]}")
print(f"Email: {users[1][1]}")
print(f"Mobile No: {users[1][2]}")

# c) Print mobile number of 3rd user
print("\n--- Mobile number of 3rd User ---")
print(f"Mobile No: {users[2][2]}")
```

```python
class takeUserDetails:

  def __init__(self,n,e,mn):     # __init__(c1,n,e,mn)           __init__(c2,n,e,mn)
    self.name = n                # c1.name = "John"               c2.name = "Alex"
    self.email = e               # c1.email = "j@gmail.com"       c2.email = "al@gmail.com"
    self.mobno = mn              # c1.mobno = 603564543           c2.mobno = 601424232

  def printDetails(self):                             # printDetails(c1)
    print(self.name,self.email,self.mobno)            # c1.name  c1.email  c1.mobno


c1 = takeUserDetails("John","j@gmail.com",603564543)
c2 = takeUserDetails("Alex","al@gmail.com",601424232)
c3 = takeUserDetails("Tina","tn@gmail.com",646362721)

# Printing all information
print('---- Information Of All Users ----')
c1.printDetails()
c2.printDetails()
c3.printDetails()

# Print the second user information
print('\n\n---- Information Of Second User----')
c2.printDetails()

# Print the mobno of 3rd user
print('\n\n---- Third User Mob No----')
print(c3.mobno)
```

```python
# Exception handling in Python is a mechanism that allows you to handle runtime errors,
#  or exceptions, in a controlled way, rather than letting your program crash unexpectedly.
# Python provides a way to catch errors, respond to them, and allow your program to continue
# running smoothly even if something goes wrong.

# try except finally
# try : keep that block of code inside try , which might give you an error
# except : catch that error

# Without Exception Handling

val1 = int(input('Enter First Value : '))
val2 = int(input('Enter Second Value : '))
res = val1 / val2
print('After Division :',res)
print('Rest Of The Code')
```

```python
# With Exception Handling

try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except:
  print('Something Went Wrong')
print('Rest Of The Code')
```

```python
try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except Exception as e:
  print('Error :',e)
print('Rest Of The Code')
```

```python
try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except ZeroDivisionError as e:
  print('Error :',e)
print('Rest Of The Code')
```

```python
# define multiple exceptions
# but only one exception will work at a same time

try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except (ZeroDivisionError,ValueError) as e:
  print('Error :',e)
print('Rest Of The Code')
```

```python
# finally keyword
# it will execute always

# Case 1 : when there is no exception

try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except Exception as e:
  print('Error :',e)
finally:
  print('I will execute always')
print('Rest Of The Code')
```

```python
# Case 2 : when there is an exception

try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except Exception as e:
  print('Error :',e)
finally:
  print('I will execute always')
print('Rest Of The Code')
```

```python
# Case 3 : when there is an exception but not handeled properly
try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except ModuleNotFoundError as e:
  print('Error :',e)
finally:
  print('I will execute always')
print('Rest Of The Code')
```
