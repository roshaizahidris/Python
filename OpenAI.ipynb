```python
# Task 1:
# Sentiment Analysis
# Enter User Review : The movie was disaster
# Output : negative üòû

# Response:
review = input("Enter User Review: ")

prompt = f"""
Classify the sentiment of this review as Positive, Negative, or Neutral.
Return only one word.

Review: "{review}"
"""

response = client.responses.create(
    model="gpt-5.2",
    input=prompt
)

sentiment = response.output_text.strip().lower()

# Add emoji
emoji_map = {
    "positive": "üòä",
    "negative": "üòû",
    "neutral": "üòê"
}

print("Sentiment:", sentiment, emoji_map.get(sentiment, ""))
```

```python
# Task 4
# Language Translator
# Input : malay
# insert text in malay   keluar
# to_language : English
# output : exit

# Response:

def translate_text(text, target_language="English"):
    prompt = f"Translate the following text to {target_language}:\n\n{text}"
    
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful language translator."},
            {"role": "user", "content": prompt}
        ]
    )
    
    # Get the translated text
    return response.choices[0].message.content

# Input in Malay
malay_text = input("Insert text in Malay: ")

# Translate to English
english_text = translate_text(malay_text, target_language="English")

# Output
print("Translated to English:", english_text)
```
